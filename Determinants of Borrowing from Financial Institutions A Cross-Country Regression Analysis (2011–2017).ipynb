```python
# Importing Libraries

import pandas as pd
import statsmodels.api as sm
```

```python
# Loading the File

file_path = "C:/Users/Emmanuel Cerrer/Desktop/Portfolio Projects/Topline Macroeconomic Dashboard (Pivot & Trend Summary).xlsx"
df = pd.read_excel(file_path, sheet_name="Cleaned - Data Set")
```

```python
# Pivot Wider

pivot_df = df.pivot_table(
    index=["Country Name", "Year"],
    columns="Metric",
    values="Value"
).reset_index()
```

```python
# Keeping Relevant Variables (Columns With Usable Data)

cols = [
    "Country Name", "Year",
    "Deposit accounts per 1,000 adults",
    "Access to internet (% age 15+)",
    "E-money accounts per 1,000 adults",
    "Borrowed from a financial institution or used a credit card (% age 15+)"
]
data = pivot_df[cols]
```

```python
# Filling Missing Values with Column Means

data_imputed = data.fillna(data.mean(numeric_only=True))
print("✅ Data shape after imputation:", data_imputed.shape)
```

```python
# Defining Dependent (y) and Independent (X) Variables

y = data_imputed["Borrowed from a financial institution or used a credit card (% age 15+)"]
X = data_imputed[
    [
        "Deposit accounts per 1,000 adults",
        "Access to internet (% age 15+)",
        "E-money accounts per 1,000 adults"
    ]
]
```

```python
# Adding a Constant (For the Intercept)

X = sm.add_constant(X)
```

```python
# Fitting the Regression Model

model = sm.OLS(y, X).fit()
```

```python
# Summary Display

print(model.summary())
```
